```python
data = dataset.sample(frac=0.9, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *
telecom = setup(data,target = 'Churn',
                max_encoding_ohe = 100,
                fold_strategy = 'kfold',
                fold = 5,
                data_split_stratify = True ,
                transformation = False,
                train_size = 0.8,
                ignore_features = ['customerID'],
                ordinal_features = {'Contract' : ['Month-to-month' ,'One year', 'Two year']},
                normalize = True,
                normalize_method ='minmax',
                low_variance_threshold = 0.1,
                numeric_imputation ='mean',
                categorical_imputation='mode',
                session_id = 43)
```

```python
# compare all models
best_model = compare_models(sort = 'Accuracy')
```

```python
#To view all available algorithms
models()
```

```python
# to create a KNN model with k=5
knn=create_model('knn',n_neighbors = 5 , metric = 'manhattan')
```

```python
#Graphical plot
evaluate_model(knn)
```

```python
#Graphical plot
evaluate_model(best_model)
```

```python
# Tune the KNN model
tuned_knn = tune_model(knn)
```
